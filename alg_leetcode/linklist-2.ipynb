```python
# Definition for singly-linked list.
# class ListNode(object):
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next


class Solution(object):
    def addTwoNumbers(self,l1, l2):
        def trans(x):
            result_list = ''
            result = x
            while result:
                result_list = result_list+str(result.val)
                result = result.next
            return int(result_list[::-1])
        def append_list_node(list_node,new_val):
            append_node = ListNode(new_val)
            while list_node.next != None:
                list_node = list_node.next
            list_node.next = append_node
        def anti_trans(x):
            x_init=str(x)
            new = ListNode(int(x_init[0]))
            for i in range(1,len(x_init)):
                append_list_node(new,int(x_init[i]))
            return new
        l1 = trans(l1)
        l2 = trans(l2)
        l = l1+l2
        l = str(l)
        l = l[::-1]
        l = anti_trans(l)
        return l
    
```

```python
class Solution(object):
    def addTwoNumbers(self, l1, l2):
        """
        :type l1: ListNode
        :type l2: ListNode
        :rtype: ListNode
        """
        ptr1 = l1 # pointer to traverse l1
        ptr2 = l2 # pointer to traverse l2
        head = l1 # pointer to the head of the result linked list, initially set to the head of l1
        sum = 0
        carry = 0
        
        # traverse both linked lists simultaneously until at least one of them is not empty
        while ptr1 or ptr2:
            num1 = ptr1.val if ptr1 else 0 # if ptr1 is not null, set num1 to its value, else set it to 0
            num2 = ptr2.val if ptr2 else 0 # if ptr2 is not null, set num2 to its value, else set it to 0
            ptr1 = ptr1.next if ptr1 else None # if ptr1 is not null, move ptr1 to the next node
            ptr2 = ptr2.next if ptr2 else None # if ptr2 is not null, move ptr2 to the next node
            
            sum = num1 + num2 + carry # calculate sum
            carry = sum / 10 # calculate carry
            sum = sum % 10 # calculate sum
            
            n = ListNode(sum) # create a new node with sum as its value
            l1.next = n
            l1 = l1.next
            num1 = num2 = sum = 0
        
        if carry:
            n = ListNode(carry)
            l1.next = n
            l1 = l1.next
        
        return head.next
```
